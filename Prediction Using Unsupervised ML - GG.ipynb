### **K Means clustering**
In this task we will predict the number of clusters what should be the optimum number of clusters for K Means? How does one determine the value of K?

```python
#First we import the libraries that will be used throughout the code
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans as kms
from sklearn import datasets as ds
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
%matplotlib inline
```

### Data Wrangling Process

#### Data Collection

```python
#Read the dataset
iris_ds = ds.load_iris()
print("Data imported successfully")
X = iris_ds.data
X[0:10]
```

### Data Selection, Cleaning & Transforming

**Since the data is clean, it does not require any changes or any transformations**

#### Machine Learning Process

```python
#K Means clustering - Finding the right K
weights=[]
for i in range(1,11):
    km = kms(n_clusters= i)
    km.fit(X)
    weights.append(km.inertia_)

plt.plot(range(1, 11),weights)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') # Within cluster sum of squares
plt.show()
```

Because after the value 3, we can see that the sum of squares does not decrese significantly and the elbow occurs here, we would use the value of K as 3 i.e. the no. of clusters = 3

```python
km = kms(n_clusters=3)
km.fit_predict(X)
labels = km.labels_
```

### Plotting of graphs

```python
fig = plt.figure(1, figsize=(7,7))
ax = Axes3D(fig, rect=[0, 0, 0.95, 1], elev=48, azim=134)
ax.scatter(X[:, 3], X[:, 0], X[:, 2],
          c=labels.astype(np.float), edgecolor="k", s=50)
ax.set_xlabel("Petal width")
ax.set_ylabel("Sepal length")
ax.set_zlabel("Petal length")
plt.title("K Means", fontsize=14)
```
